```python
import pandas as pd
import numpy as np
```

```python
df_london = pd.read_csv(
    "/Users/jonahwitte/Desktop/IS_477_Group_Project/Data Folder/London Bloomsbury_AllPollutants_Daily_2024.csv",
    skiprows=7,
)
df_london = df_london.replace("no data", pd.NA)
remove_status = [
    "Date",
    "Nitrogen dioxide",
    "Ozone",
    "PM10 particulate matter (Hourly measured)",
    "PM2.5 particulate matter (Hourly measured)",
    "Sulphur dioxide"
]
df_london = df_london[remove_status]
df_london.head()

#df_deaths = pd.read_csv("/Users/jonahwitte/Desktop/IS_477_Group_Project/Data Folder/Ambient_Air_Pollution_deaths.csv")
#df_deaths.head()
```

```python
pollutant_cols = [
    "Nitrogen dioxide",
    "Ozone",
    "PM10 particulate matter (Hourly measured)",
    "PM2.5 particulate matter (Hourly measured)",
    "Sulphur dioxide"
]
df_london[pollutant_cols] = df_london[pollutant_cols].apply(pd.to_numeric)
```

```python
df_london["Date"] = pd.to_datetime(df_london["Date"])
df_london["month"] = df_london["Date"].dt.month
#df_london.head()
df_london.groupby("month")[pollutant_cols].mean()
```
